```python
from sklearn import linear_model
LR_model = linear_model.LinearRegression()

x = np.asanyarray(df[['TV']])
y = np.asanyarray(df[['Sales']])

LR_model.fit(x, y)

theta1 = LR_model.coef_[0][0]
theta0 = LR_model.intercept_[0]
print(theta0)
print(theta1)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='TV', y='Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='TV', y='Sales')
plt.plot(x, theta0+theta1*x, '-r')
plt.show()
```

```python
def linear(x):
    return theta0+theta1*x
```

```python
df_new = df.sample(50)
df_new.drop(['Radio', 'Newspaper'], axis=1, inplace=True)
df_new
```

```python
df_new['Sales_pred'] = linear(df['TV'])
df_new
```

```python
np.sqrt((1/df_new.shape[0])*((df_new['Sales_pred']-df_new['Sales'])**2).sum())
```
